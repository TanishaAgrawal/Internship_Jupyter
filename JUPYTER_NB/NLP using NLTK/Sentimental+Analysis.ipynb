```python
tweetData = pd.read_csv('tweets.csv')
```

```python
tweetData = tweetData[['tweet_text','tweet_sentiment_value']]
#tweetData = tweetData.head(20)
```

```python
sid = SentimentIntensityAnalyzer()
i = 0
tweetData['Score'] = ''
tweetData['Prediction'] = ''
for sentence in tweetData['tweet_text']:
    print(sentence)
    ss = sid.polarity_scores(sentence)
    comp = ss['compound']
    tweetData['Score'][i] = comp
    if comp>=0.01:
        tweetData['Prediction'][i] = 2 
    elif comp<=-0.01:
        tweetData['Prediction'][i] = 0
    else:
        tweetData['Prediction'][i] = 1
    for k in ss:
        print('{0}: {1}, '.format(k, ss[k]), end='')
        print()
    i = i + 1
    
```

```python
tweetData['tweet_sentiment_value'] = tweetData['tweet_sentiment_value'].astype(int)
tweetData[(tweetData.tweet_sentiment_value == 0)]
```

```python
tp = len(tweetData[tweetData.tweet_sentiment_value == tweetData.Prediction])
print(tp)
```

```python
tp = len(tweetData[(tweetData.tweet_sentiment_value == 2) & (tweetData.Prediction == 2)])
tn = len(tweetData[(tweetData.tweet_sentiment_value == 0) & (tweetData.Prediction == 0)])
fp = len(tweetData[(tweetData.tweet_sentiment_value == 2) & (tweetData.Prediction != 2)])
fn = len(tweetData[(tweetData.tweet_sentiment_value == 0) & (tweetData.Prediction != 0)])
tne= len(tweetData[(tweetData.tweet_sentiment_value == 1) & (tweetData.Prediction == 1)])
fne= len(tweetData[(tweetData.tweet_sentiment_value == 1) & (tweetData.Prediction != 1)])
print ("True Positives:", tp)
print ("True Negatives:", tn)
print ("False Positives:", fp)
print ("False Negatives:", fn)
print ("True Neutral:", tne)
print ("False Neutral:", fne)
print ("Total", len(tweetData))
```
